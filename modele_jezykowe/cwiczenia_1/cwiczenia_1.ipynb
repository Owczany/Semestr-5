### Zadanie 5 – Generacja dokładnie jednego wyrazu

Aby efektywnie wygenerować **dokładnie jeden wyraz** przy użyciu podstawowego interfejsu modelu językowego (czyli funkcji przewidującej kolejne tokeny na podstawie zadanego prefiksu), można zastosować następującą procedurę:

Zakładamy, że model działa na poziomie tokenów (np. sub-słów) i generuje kolejne tokeny jeden po drugim. Dla przykładu, jeśli podamy modelowi prompt:

> **„Ala idzie do”**

to model może próbować wygenerować pełne zdanie, np. **„sklepu po zakupy”**. Naszym celem jest jednak przerwać generację **dokładnie po jednym pełnym wyrazie**, czyli np. uzyskać tylko **„sklepu”**.

Ponieważ większość nowoczesnych modeli językowych (np. opartych na BPE) koduje tokeny w taki sposób, że tokeny reprezentujące początek nowego słowa zaczynają się od spacji (np. `" sklep"`), możemy to wykorzystać.

---

#### **Procedura:**
1. Podajemy modelowi prefiks i prosimy o generację tokenów krok po kroku.
2. Oczekujemy na pierwszy wygenerowany token, który zaczyna się od spacji – oznacza to początek nowego słowa.
3. Zbieramy ten token oraz wszystkie kolejne tokeny, które **nie zaczynają się od spacji** (czyli są kontynuacją tego samego słowa).
4. Zatrzymujemy generację, gdy napotkamy token zaczynający się od spacji – to oznacza początek kolejnego słowa.
5. Łączymy zebrane tokeny (np. `tokenizer.decode()`) i zwracamy dokładnie jeden wyraz.

---

#### **Przykład:**
- Model generuje tokeny: `" skle"`, `"pu"`, `" i"`
- Zbieramy: `" skle"` + `"pu"` → daje słowo **„sklepu”**
- Przerywamy przed `" i"` – bo to początek nowego wyrazu

---

Dzięki temu otrzymujemy precyzyjnie **jeden pełny wyraz**, co spełnia założenia zadania.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# Wybór urządzenia
device = "cuda" if torch.cuda.is_available() else ("mps" if torch.backends.mps.is_available() else "cpu")

# Wczytaj tokenizer i model
model_name = "Eryk-Mazus/polka-1.1b-chat"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

# Pipeline do generowania
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, device=0 if device=="cuda" else -1)

# Prompt do tłumaczenia
text_en = "The woman is traveling by plane, train or ship."
prompt = f"Przetłumacz na język polski: {text_en}\nOdpowiedź:"

print('Model Loaded')

# Generowanie
output = generator(prompt, max_new_tokens=50, do_sample=False)[0]["generated_text"]

# Wyciągnij samą odpowiedź
translated = output.split("Odpowiedź:")[-1].strip()
print("Polskie tłumaczenie:", translated)
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline

MODEL_NAME = "sdadas/mt5-base-translator-en-pl"
DEVICE = "cuda" if torch.cuda.is_available() else ("mps" if torch.backends.mps.is_available() else "cpu")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME).to(DEVICE)

translator = pipeline(
    "translation_en_to_pl",
    model=model,
    tokenizer=tokenizer,
    device=0 if DEVICE == "cuda" else -1
)

result = translator("The weather is nice today.")
print(result[0]['translation_text'])
```
