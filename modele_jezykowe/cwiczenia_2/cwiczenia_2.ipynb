W tym zadaniu można wykorzystać fakt, że model językowy generując tekst, przewiduje listę możliwych kolejnych tokenów z przypisanymi prawdopodobieństwami.
Osoba kodująca wiadomość może przekazać ciąg binarny (np. 00010001) i w trakcie generacji wybierać tokeny zgodnie z jego bitami:
jeśli bit = 0 → wybieramy token o parzystym indeksie z listy przewidywanych tokenów,
jeśli bit = 1 → wybieramy token o nieparzystym indeksie.
Odbiorca, znając ten sam model i procedurę, może odtworzyć ten sam rozkład prawdopodobieństwa tokenów w każdym kroku i sprawdzić, czy w danej pozycji został wybrany token parzysty czy nieparzysty. W ten sposób odczyta ukryty ciąg binarny, a sam tekst pozostaje w pełni naturalny dla osób postronnych.

Zasada Kerckhoffsa mówi, że bezpieczeństwo systemu szyfrowania nie powinno zależeć od tajności algorytmu, lecz wyłącznie od tajności klucza.
W naszym poprzednim rozwiązaniu ta zasada nie jest spełniona, ponieważ jeśli ktoś zna sposób kodowania (np. że token o parzystym indeksie oznacza 0, a o nieparzystym 1), to bez problemu odczyta ukrytą wiadomość.
Aby to naprawić, można wprowadzić tajny klucz, który zmienia regułę interpretacji tokenów w zależności od pozycji.
Przykładowo klucz może być tablicą lub mapą określającą, jak na danej pozycji interpretować parzystość tokenu:

klucz = {

    0: {"parzyste": 0, "nieparzyste": 1},

    1: {"parzyste": 1, "nieparzyste": 0},

    2: {"parzyste": 0, "nieparzyste": 1},

    ...
    
}

Podczas kodowania dla pozycji i wybieramy regułę zgodnie z klucz[i % n].
Odbiorca, znając ten sam klucz, może poprawnie zdekodować wiadomość, natomiast osoba trzecia, nieznająca klucza, nie będzie w stanie rozróżnić znaczenia poszczególnych tokenów.
Dzięki temu system spełnia Zasadę Kerckhoffsa – jego bezpieczeństwo zależy tylko od tajnego klucza, a nie od ukrywania samego sposobu działania.

## Cechy charakterystyczne tekstu
Tekst, taki jak „Biega, krzyczy pan Hilary”, charakteryzuje się:
- regularnym rytmem – np. 8 sylab w każdym wersie,
- rymami (najczęściej układ AABB lub ABAB),
- prostą, rytmiczną składnią,
- regularną interpunkcją i puentą na końcu,
- lekką, często humorystyczną tematyką.

## Proponowany algorytm generacji
Aby wykorzystać model językowy (np. Papugę) do tworzenia takich wierszyków, można zaprojektować następujący algorytm:

1. **Określenie struktury wiersza** – ustalamy liczbę wersów, układ rymów (np. AABB) oraz liczbę sylab w wersie (np. 8).
2. **Generacja wersów** – model generuje kolejne słowa, dopóki liczba sylab nie osiągnie celu. Jeśli liczba sylab przekracza limit, generacja jest zatrzymywana lub korygowana.
3. **Kontrola rymów** – dla wersów mających się rymować, końcowe słowo dobierane jest tak, by miało podobną końcówkę fonetyczną do poprzedniego wersu. Można to osiągnąć przez re-ranking propozycji modelu według podobieństwa rymu.
4. **Ograniczenie przeszukiwania** – zamiast generować wszystkie możliwości, stosujemy algorytm *beam search* o małej szerokości (np. 5), zachowując kilka najlepszych kandydatów.
5. **Ocena wyników** – wybieramy wierszyk najlepiej spełniający kryteria rytmu, rymu i płynności językowej.

## Przykład działania
Dla schematu 4 wersy, 8 sylab, układ AABB:
```
Biega, krzyczy pan Hilary,
Gdzie się podział jego szalik stary,
Szukaj w spodniach albo w bucie,
Wszystko znajdzie w wielkim skrócie.
```

## Podsumowanie
Taki tekst jest regularny, rytmiczny i rymowany. Model językowy może generować podobne wiersze, jeśli proces generacji kontroluje liczbę sylab i końcówki słów, a wybór najlepszej wersji odbywa się przez re-ranking lub ograniczone przeszukiwanie (beam search).
